## 공분산 행렬(Covariance Matrix)

### 공분산 구하는 법
열 벡터값을 가지는 확률변수 X와 Y가 각각 $\mu$ 와 $\nu$라는 기댓값을 가질 때

공분산 $2 x 2$ 행렬은 아래와 같다.

$Cov(X,Y) = E [\left\{(X-\mu)(Y- \nu) \right\} $]

공분산은 2개의 확률변수간의 선형의존성을 나타내는 척도로도 쓰인다.

즉, 위에서 봤을 때 X와 Y의 선형관계가 얼마나 있는가를 확인할 수 있다.

### 기하학적 의미

공분산 행렬의 각 원소의 의미

<img src = 'https://raw.githubusercontent.com/angeloyeo/angeloyeo.github.io/master/pics/2019-07-27_PCA/pics_mtx1.png'>

- 1행 1열은 x의 분산
- 1행 2열과 2행 1열은 x와 y의 공분산
- 2행 2열은 y의 분산

### 수식적 의미

여기서 짚고 넘어가야 하는 부분은 분산을 구하는 법과 공분산을 구하는 법이다.

분산을 구하는 공식

Var(X) = $E[(X - \mu)^2]$

공분산을 구하는 공식

$Cov(X,Y) = E [\left\{(X-\mu)(Y- \nu) \right\} $]

## Linear Model VS Linear Mixed Model

### Linear Model

$Y = \beta X + \epsilon$

$E[Y] = X\beta$

where
\begin{cases}
E(e) = 0\\
Var(e) = R\\
\end{cases}

### Linear Mixed Effect Model

$Y = X \beta + Z \mu + \epsilon $

where
\begin{cases}
E(e) = 0\\
Var(e) = R\\
Var(Y|u) = R\\
E(u) = 0\\
Var(u) = D
\end{cases}

## Linear Mixed Effect Model tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('C:/Users/lg/캐글데이터분석/House Prices/train.csv')
data.head()
```

```python
# 데이터 분포 확인해 보기
plt.figure(figsize =(22,8))
sns.distplot(data['SalePrice'])
```

```python
# 종속변수를 데이터의 중간으로 맞춰주도록 하자.
data['SalePrice'] = data['SalePrice'].apply(lambda x:np.log1p(x))
```

```python
plt.figure(figsize = (22,8))
sns.distplot(data['SalePrice'])
```

```python
# 데이터 결측치 확인해보기
print(data.isnull().sum())

# LMM의 데이터를 분석하기 위해서 하는 tutorial임으로 결측치가 있는 column을 전부 제거한다.
data = data.dropna(axis = 1)
```

```python
data.columns
```

```python
# 데이터 분포 확인해보기
plt.figure(figsize =(22,8))
sns.distplot(data['LotArea'])
```

```python
data['LotArea'] = data['LotArea'].apply(lambda x: np.log1p(x))

# standard scaler를 한 이후 데이터 분포 확인해보기
plt.figure(figsize = (22,8))
sns.distplot(data['LotArea'])
```

```python
plt.figure(figsize = (16,8))
sns.lmplot(data = data, x = 'LotArea', y = 'SalePrice')
```

```python
# Linear Regression으로 rmse가 얼마나 나오나 확인해보자.
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept = True)
X = data['LotArea']
y = data['SalePrice']

model.fit(X[:,np.newaxis],y)

xfit = np.linspace(6, 13, 1000)
yfit = model.predict(xfit[:,np.newaxis])
plt.scatter(X,y)
plt.plot(xfit, yfit)
```

```python
y_predict = model.predict(X.values.reshape(-1, 1))
RMSE = np.sqrt(((y - y_predict)**2).values.mean())

results = pd.DataFrame()
results['Method'] = ['Linear Regression']
results['RMSE'] = RMSE
results
```

```python
data.columns
```

```python
# 각 값마다 SalePrice에 차이가 있다고 판단한다.
sns.barplot(data = data, x = 'MSSubClass', y = 'SalePrice')
```

```python
# 각 MSSubClass별로 기울기나 절편의 값이 전부 다르다.
sns.lmplot(data = data, x = 'LotArea', y = 'SalePrice', col = 'MSSubClass', col_wrap = 3, height = 5)
```

```python
data_new = pd.concat([data['LotArea'], pd.get_dummies(data['MSSubClass'])], axis = 1)
data_new.head()
```

```python
from yellowbrick.regressor import ResidualsPlot
model = LinearRegression(fit_intercept = True)

X = data_new
y = data['SalePrice']

model.fit(X, y)

visualizer = ResidualsPlot(model = model)
visualizer.fit(X,y)
visualizer.poof()
```

```python
y_predict = model.predict(X)
RMSE = np.sqrt(((y - y_predict)**2).values.mean())
results.loc[1] = ['Fixed', RMSE]
results
```

```python
pd.DataFrame.from_records(list(zip(np.concatenate((['LotArea'],data['MSSubClass'].unique())), model.coef_)))
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# smf.mixedlm을 사용하여 LotArea, 절편, Group 분산을 파라미터로 넣는다.
md = smf.mixedlm('SalePrice ~ LotArea', data, groups = data['MSSubClass'])
mdf = md.fit()
print(mdf.summary())
```

```python
# 위의 summary를 가지고 데이터를 가지고 y_predict를 만든다.
y_predict = mdf.fittedvalues
RMSE = np.sqrt(((y-y_predict)**2).values.mean())
results.loc[2] = ['Mixed',RMSE]
results

# 그러나 데이터의 RMSE가 Fixed와 Mixed에서 차이가 없다.
# 이유는 LotArea의 파라미터와 Group 분산의 파라미터만 들어갔기 때문이다.
# 그러면 LotArea와 Group의 공분산과 LotArea의 분산을 넣어준다면 어떻게 변하는지 확인해 보자.
```

```python
md = smf.mixedlm('SalePrice ~ LotArea', data, groups = data['MSSubClass'], re_formula = '~LotArea')
mdf = md.fit()
print(mdf.summary())
```

```python
# RMSE가 조금은 줄어드는 것을 볼 수 있다.
y_predict = mdf.fittedvalues
RMSE = np.sqrt(((y-y_predict)**2).values.mean())
results.loc[3] = ['Mixed_Random_Slopes', RMSE]
results
```
